```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

class linkedlist:
  def __init__(self):
    self.head = None

  def inserthead(self,newdata):
    tempnode = Node(newdata)
    tempnode.next = self.head.next
    tempnode = self.head

  def insertMid(self,Midnode,newdata):
    tempnode = Node(newdata)
    tempnode.next = Midnode.next
    Midnode.next = tempnode

  def remove(self,removedata):
    temp = self.head
    while temp is not None:
      if temp.data == removedata:
        break
      previousNode = temp
      temp = temp.next
    previousNode.next = temp.next
    
  def show(self):
    temp = self.head
    while temp is not None:
      print(temp.data)
      temp = temp.next

N1 = linkedlist()
N1.head = Node("Rathna")
N2 = Node("Rohini")
N3 = Node("Suba")
N4 = Node("Hari")
N5 = Node("Geetha")

N1.head.next = N2
N2.next = N3
N3.next = N4
N4.next = N5


print("Before Insert")
N1.show()
print('***************************************')

N1.inserthead('Lingam')
print("After Insert Head Node")
N1.show()
print('***************************************')

print("After Insert Mid Node")
N1.insertMid(N3,'Vivek')
N1.show()
print('***************************************')

print("After remove the Data")
N1.remove('Geetha') 
N1.show()
print('***************************************')
```
